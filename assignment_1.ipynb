## Hello, Graph World

This notebook demonstrates the creation and visualization of a basic graph using the Krackhardt kite structure. The graph is constructed with NetworkX, a Python library for the creation, manipulation.

### Objective
To define nodes representing individuals and edges representing their relationships, then visualize the resulting network.



```python
#importing the libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# create an empty graph
G = nx.Graph()

# adding nodes to the graph
nodes = ["Andre", "Beverly", "Carol", "Diane", "Ed", "Fernando", "Garth", "Heather", "Ike", "Jane"]
G.add_nodes_from(nodes)

# print the nodes for verification
print("Nodes in the graph:")
print(G.nodes())
```

```python
# adding edges based on the Krackhardt kite diagram
edges = [
    ("Andre", "Carol"),
    ("Andre", "Diane"),
    ("Andre", "Beverly"),
    ("Beverly", "Diane"),
    ("Beverly", "Ed"),
    ("Carol", "Fernando"),
    ("Carol", "Diane"),
    ("Diane", "Garth"),
    ("Diane", "Fernando"),
    ("Ed", "Garth"),
    ("Fernando", "Heather"),
    ("Garth", "Heather"),
    ("Heather", "Ike"),
    ("Ike", "Jane")
]
G.add_edges_from(edges)

# observe edges are in the graph
print("edges in the graph:")
print(G.edges())
```


## Graph Creation
Define nodes for each individual and edges to represent their connections.

```python
# creating the graph
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=12)
plt.title("Basic Graph Visualization")
plt.show()
```
